```python
from google.colab.patches import cv2_imshow
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import math

app=cv2.imread("/content/apple.jpg",cv2.IMREAD_GRAYSCALE)
org=cv2.imread("/content/orange.jpg",cv2.IMREAD_GRAYSCALE)
app2=cv2.imread("/content/apple.jpg",cv2.IMREAD_GRAYSCALE)

cv2_imshow(app)
cv2_imshow(org)
cv2_imshow(app2)
```

```python
mix=app.copy()
for i in range(app.shape[0]):
  for j in range (app.shape[1]):
    mix.itemset((i,j),(0.3*app[i][j]+0.3*org[i][j]+0.3*app2[i][j]))

cv2_imshow(mix)
```
